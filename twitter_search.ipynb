```python
import requests, json
from random import randrange
from datetime import datetime
from TwitterAPI import TwitterAPI

# Read in API Keys
f = open('api.key', 'r')
access_token = f.readline().strip('\n')
access_token_secret = f.readline().strip('\n')
consumer_key = f.readline().strip('\n')
consumer_secret = f.readline().strip('\n')
f.close()
```

## Standard Search

```python
SEARCH_TERM = 'marijuana -Cop -smelt'

api = TwitterAPI(consumer_key, 
                 consumer_secret,
                 access_token,
                 access_token_secret)

r = api.request('search/tweets', {'q': SEARCH_TERM})

for item in r:
    print(item['text'] if 'text' in item else item)
    
#        f = open('data/export' + '2' + '.json', 'a+')
 #       f.write(data)
  #      f.close()

print('\nQUOTA: %s' % r.get_quota())
```

## Premium Search: last 30 days

```python
SEARCH_TERM = 'pizza'
PRODUCT = '30day'
LABEL = '30days'

r = api.request('tweets/search/%s/:%s' % (PRODUCT, LABEL), 
                {'query':SEARCH_TERM})

for item in r:
    print(item['text'] if 'text' in item else item)

print('\nQUOTA: %s' % r.get_quota())
```
